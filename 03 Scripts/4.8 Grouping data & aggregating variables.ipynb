### Step 4. Create loyality flag for customer based on numbers of order

```python
# Aggregating data with transform function

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# checking aggregated data
ords_prods_merge.head(10)
```

```python
# creating loyality flag with loc
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# checkung value count for loyalty customer flag
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

all flags add up to 32404859, total rows in data.

```python
ords_prods_merge.head()
```

### Step 5. Find the difference between the spending habits of three types of customer based on loyalty

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean','min','max']})
```

Average expenditure in all three category is almost same,$7.7 - 7.8. And so is the minimun and maximum
We can say, prices of the product purchased by all three type of customers is more or less same.


### Step 6. Create a spending flag for each user based on the average price across all their orders

```python
# Aggregating data with transform function

ords_prods_merge['average_spending']= ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge[['user_id','average_spending']].head(100)
```

```python
# Creating the spending label
ords_prods_merge.loc[ords_prods_merge['average_spending'] < 10 , 'spending_label'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['average_spending'] >= 10, 'spending_label'] = 'High spender'
```

```python
# Counting the values 
ords_prods_merge['spending_label'].value_counts(dropna = False)
```

99% of customers are in low spending flag.

```python
# checking relevant columns
ords_prods_merge[['user_id','average_spending','spending_label']].head(100)
```

### Step 7. create an order frequency flag

```python
# aggregating data with transform function
ords_prods_merge['days_btwn_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge[['user_id','days_since_prior_order','days_btwn_orders']].head()
```

```python
# Creating 'Order_frequency' flag
ords_prods_merge.loc[ords_prods_merge['days_btwn_orders'] > 20, 'order_frequency'] = 'Non frequent customer'
ords_prods_merge.loc[(ords_prods_merge['days_btwn_orders'] <= 20) & (ords_prods_merge['days_btwn_orders'] >10), 'order_frequency'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['days_btwn_orders'] <= 10, 'order_frequency'] = 'Frequent customer'
```

```python
# counting values
ords_prods_merge['order_frequency'].value_counts(dropna= False)
```

70 % customers order frequently,
21% orders regularly,
and 8 % customers orders rarely
